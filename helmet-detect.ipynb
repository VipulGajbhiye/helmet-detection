```python
import io
from PIL import Image
```

```python
import ipywidgets as widgets
```

```python
image_dir = "data/images/"
```

```python
from detect import detect
```

```python
btn_image = widgets.FileUpload()
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
```

```python
out_img = widgets.Output()
out_img.clear_output()
```

```python
iou_thres = widgets.FloatSlider(value=0.45, min=0.05, max=0.95, step=0.05, description="IoU threshould")
conf_thres = widgets.FloatSlider(value=0.25, min=0, max=1, step=0.05, description="Confidence threshould")
```

```python
btn_run = widgets.Button(description="Detect")
```

```python
labl_pred = widgets.HTML()#widgets.Label()
```

```python
cust_pred = widgets.Checkbox(
    value=True,
    description='Helmet & person',
    disabled=False,
    tooltip='Helmet detection or default COCO classes',
    indent=False
)
```

```python
wts_dict = {True:'weights/best.pt', False: 'weights/yolov5s.pt'}
```

```python
def on_click_classify(change):
    thumb_size = 460
    img = btn_image.data[-1] # bytes code
    img = Image.open(io.BytesIO(img))
    img.save(image_dir +'img.jpg')
    img.thumbnail((thumb_size,thumb_size))
    img.save(image_dir +"img_thumb.jpg")
    out_pl.clear_output()
    with out_pl:
        img2=Image.open(image_dir +"img_thumb.jpg")
        display(img2)
        #img.thumbnail((256,256)))
    #pred, pred_idx, probs = learn_inf.predict(img)
    #lbl_pred.value = f"\t Prediction: {pred.capitalize()} bear with probability {round(probs[pred_idx].item()*100,2)}%."
    
    s_im, s, t = detect(source=image_dir +"img.jpg", weights=wts_dict[cust_pred.value], iou_thres=iou_thres.value, conf_thres=conf_thres.value)
    text = f"      Predicted: {s} in {round(t,2)} s."
    labl_pred.value = f"<h4><font color='blue'>{text}</h4>"
    img = Image.open('yolov5s/img.jpg')
    img.thumbnail((thumb_size,thumb_size))
    img.save(image_dir +"img_thumb.jpg")
    out_img.clear_output()
    with out_img:
        img2 = Image.open(image_dir +'img_thumb.jpg')
        display(img2)

def on_upload_display(change):
    img = btn_image.data[-1] # bytes code
    img = Image.open(io.BytesIO(img))
    img.save('img_orig.jpg')
    out_pl.clear_output()
    with out_pl:
        display(img)
        #display(img.thumbnail((256,256)))
```

```python
btn_run.on_click(on_click_classify)
#btn_image.on_change(on_upload_display)
```

```python
instruction_labl = widgets.Label("Upload your image")
```

```python
input_widgets = widgets.VBox([instruction_labl, btn_image, cust_pred, iou_thres, conf_thres, widgets.HBox([btn_run, labl_pred])])
```

```python
output_widgets = widgets.HBox([out_pl, out_img])
```
